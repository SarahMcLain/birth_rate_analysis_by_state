```python
import pandas as pd
import numpy as np
import matplotlib as plt
import seaborn as sns
```

```python
birth_census = pd.read_csv("data/us_births_2016_2021.csv")

birth_census
```

```python
birth_census["Education"] = birth_census["Education Level of Mother"]

birth_census_copy = birth_census.copy()
birth_census_copy.head()
```

```python
birth_census_copy.drop(['Education Level of Mother'], axis=1, inplace=True)

birth_census_copy.head()
```

```python
birth_census_copy.drop(birth_census_copy.columns[4], axis=1, inplace=True)
birth_census_copy.head()
```

```python
birth_census_copy.drop(['State'], axis=1, inplace=True)
birth_census_copy.head()
```

```python
total_births_by_state = {}

for index, row in birth_census_copy.iterrows():
    state = row['State Abbreviation']
    number_of_births = row['Number of Births']

    if state in total_births_by_state: 
        total_births_by_state[state] += number_of_births
    else:
        total_births_by_state[state] = number_of_births

print(total_births_by_state)

total_births_pd = pd.DataFrame(list(total_births_by_state.items()), columns=['State', 'Total Births'])
```

```python
total_births_pd_sorted = total_births_pd.sort_values(by='Total Births', ascending=False)
total_births_pd_sorted
```

```python
total_births_pd_sorted.dtypes
```

```python
# Group the data by state and aggregate the columns
aggregated_data = birth_census_copy.groupby(['State Abbreviation', 'Education']).agg({
    'Number of Births': 'sum',
    'Average Age of Mother (years)': 'mean',
    'Average Birth Weight (g)': 'mean'
}).reset_index()

# Display the aggregated data
print(aggregated_data)
```

```python
df_aggregated_data = pd.DataFrame(aggregated_data)

df_aggregated_data
```
